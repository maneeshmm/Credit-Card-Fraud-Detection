```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Amount'] = sc.fit_transform(pd.DataFrame(data['Amount']))
```

```python
x = data.drop(['Class'], axis= 1)
```

```python
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
(x_train,x_test,y_train,y_test) = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
classifier = {'Logistic Regression': LogisticRegression(),
              'DecisionTreeClassifier': DecisionTreeClassifier()
}
```

```python
for name, clf in classifier.items():
    print(f'=========={name}==========')
    clf.fit(x_train,y_train)
    preds = clf.predict(x_test)
    print(f'Accuracy: {accuracy_score(y_test,preds)}')
    print(f'f1_score: {f1_score(y_test,preds)}')
    print(f'precision_score: {precision_score(y_test,preds)}')
    print(f'recall_score: {recall_score(y_test,preds)}')
```

```python
#Over sampling
x = data.drop(['Class'], axis= 1)
y = data['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE()

# Apply SMOTE to resample the dataset
x_res, y_res = smote.fit_resample(x, y)
```

```python
y_res.value_counts()
```

```python
(x_train,x_test,y_train,y_test) = train_test_split(x_res,y_res, test_size = 0.2, random_state = 42)
```

```python
for name, clf in classifier.items():
    print(f'=========={name}==========')

    clf.fit(x_train,y_train)
    preds = clf.predict(x_test)
    print(f'Accuracy: {accuracy_score(y_test,preds)}')
    print(f'f1_score: {f1_score(y_test,preds)}')
    print(f'precision_score: {precision_score(y_test,preds)}')
    print(f'recall_score: {recall_score(y_test,preds)}')
```
